# Image Classification for Memes

## Outline
**1) Data**

The first step is to obtain a dataset of memes. This can be done by crawling websites like Reddit or Imgflip, or by collecting memes from social media. Once the dataset is collected, it needs to be preprocessed. This involves cleaning the data, removing duplicates, and converting the data to a format that can be used by the machine learning model.

**2) Preprocessing**

The following steps are typically involved in preprocessing meme data:
Remove any duplicate memes.
Remove any memes that are not in English.
Remove any memes that are not appropriate for all audiences.
Clean the text in the meme captions, removing any punctuation, stop words, and other unnecessary characters.
Convert the text in the meme captions to lowercase.
Split the text in the meme captions into words.

**3) Model**

There are a variety of different machine learning models that can be used for this task, but some of the most common ones include:

* Convolutional neural networks (CNNs): CNNs are a type of machine learning model that are well-suited for image classification and recognition tasks. They can be used to generate memes by learning the features of common meme templates and associating them with specific captions.
* Recurrent neural networks (RNNs): RNNs are a type of machine learning model that are well-suited for text generation tasks. They can be used to generate memes by learning the patterns in meme captions and using them to generate new captions.
* Transformers: Transformers are a type of machine learning model that are well-suited for both image and text processing tasks. They can be used to generate memes by learning the relationship between the image and the text in a meme and using that knowledge to generate new memes.

**4) Training**

Once a machine learning model has been selected, it needs to be trained on the preprocessed meme data. This involves feeding the model the data and allowing it to learn the patterns in the data. The training process can be time-consuming, but it is important to train the model for a sufficient amount of time in order to achieve good results. Once the model is trained, it can be used to generate new memes. To do this, simply provide the model with an image and it will generate a caption for the image. The caption will be generated based on the features of the image and the patterns that the model learned during training.

## 2) Preprocessing

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set the seed
tf.random.set_seed(42)

# Preprocess data (get all of the pixel values between 0 & 1, scaling/normalization)
train_datagen = ImageDataGenerator(rescale=1./255)

# Create ImageDataGenerator training and test instance with data augmentation
train_datagen_augmented = ImageDataGenerator(rescale=1/255.,
                                          rotation_range=0.5,   # rotate an image
                                          shear_range=0.2,      # shear an image
                                          zoom_range=0.2,       # zoom in on an image
                                          width_shift_range=0.3,    # move image around on x-axis
                                          height_shift_range=0.3,   # move image around on y-axis
                                          horizontal_flip=True) # flip an image

valid_datagen_augmented = ImageDataGenerator(rescale=1/255.,
                                          rotation_range=0.2,   # rotate an image
                                          shear_range=0.2,      # shear an image
                                          zoom_range=0.2,       # zoom in on an image
                                          width_shift_range=0.2,    # move image around on x-axis
                                          height_shift_range=0.3,   # move image around on y-axis
                                          horizontal_flip=False) # flip an image

test_datagen_augmented = ImageDataGenerator(rescale=1/255.,
                                          rotation_range=0.3,   # rotate an image
                                          shear_range=0.2,      # shear an image
                                          zoom_range=0.2,       # zoom in on an image
                                          width_shift_range=0.3,    # move image around on x-axis
                                          height_shift_range=0.3,   # move image around on y-axis
                                          horizontal_flip=True) # flip an image

# Setup paths to our data directories
train_dir = "Meme-Dataset/train"
test_dir = "Meme-Dataset/test"
valid_dir = "Meme-Dataset/validate"

# Import data from directories and turn it into batches
train_data = train_datagen.flow_from_directory(directory=train_dir, # test data is 80% of data, so set that to train data
                                               batch_size=24, # maybe change out batch size
                                               target_size=(224, 224),
                                               class_mode="categorical",
                                               seed=42)

train_data_augmented = train_datagen_augmented.flow_from_directory(directory=train_dir, # test data is 80% of data, so set that to train data
                                               batch_size=24, # maybe change out batch size
                                               target_size=(224, 224),
                                               class_mode="categorical",
                                               seed=42)

test_data = test_datagen_augmented.flow_from_directory(directory=test_dir,
                                               batch_size=24,
                                               target_size=(224, 224),
                                               class_mode="categorical",
                                               seed=42)

valid_data = valid_datagen_augmented.flow_from_directory(directory=valid_dir,
                                               batch_size=24,
                                               target_size=(224, 224),
                                               class_mode="categorical",
                                               seed=42)
```

```python
# Create a function to import an image and resize it to be able to be used with our model
def load_and_prep_image(filename, img_shape=224):
  """
  Reads an image from filename, turns it into a tensor and reshapes it to
  (img_shape, img_shape, color_channels).
  """
  # Read in the image
  img = tf.io.read_file(filename)
  # Decode the read file into a tensor
  img = tf.image.decode_image(img)
  # Resize the image
  img = tf.image.resize(img, [img_shape, img_shape])
  # Rescale the image (get all values between 0 and 1)
  img = img/255.
  return img
```

```python
def pred_and_plot(model, filename, class_names=class_names):
  """
  Imports an image located at filename, makes a prediction with model
  and plots the image with the predicted class as the title.
  """
  # Import the target image and preprocess it
  img = load_and_prep_image(filename)

  # Make a prediction
  pred = model.predict(tf.expand_dims(img, axis=0))

  # Add in logic for multi-class and get pred_class name
  if len(pred[0]) > 1:
    pred_class = class_names[tf.argmax(pred[0])]
  else:
    pred_class = class_names[int(tf.round(pred[0]))]

  # Plot the image and predicted class
  plt.imshow(img)
  plt.title(f"Prediction: {pred_class}")
  plt.axis(False);
```

```python
from tensorflow.keras.layers import Input, Dense, LeakyReLU, BatchNormalization, Conv2D, UpSampling2D, Reshape

image_encoder = tf.keras.Sequential([
    Input(shape=(28, 28, 1)),
    Conv2D(32, 3, padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(64, 3, padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(128, 3, padding='same', activation='relu'),
    BatchNormalization(),
])

image_decoder = tf.keras.Sequential([
    Reshape((7, 7, 128)),
    UpSampling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    BatchNormalization(),
    UpSampling2D(),
    Conv2D(32, 3, padding='same', activation='relu'),
    BatchNormalization(),
    UpSampling2D(),
    Conv2D(1, 3, padding='same', activation='sigmoid'),
])
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense

caption_encoder = tf.keras.Sequential([
    Embedding(input_dim=10000, output_dim = 128),
    LSTM(256, return_sequences=True),
    LSTM(256),
])

caption_decoder = tf.keras.Sequential([
    LSTM(512, return_sequences=True),
    LSTM(512),
    Dense(units=115, activation='softmax')
])
```

```python
image_input = Input(shape=(28, 28, 1))
caption_input = Input(shape=(20,))

encoded_image = image_encoder(image_input)
encoded_caption = caption_encoder(caption_input)

merged_representation = tf.keras.layers.Concatenate()([encoded_image, encoded_caption])

decoded_image = image_decoder(merged_representation)
decoded_caption = caption_decoder(merged_representation)

meme_model = tf.keras.Model(inputs=[image_input, caption_input], outputs=[decoded_image, decoded_caption])
```

```python
meme_model.compile(optimizer='adam', loss={'decoded_image': 'binary_crossentropy', 'decoded_caption': 'categorical_crossentropy'})
```

```python
import tensorflow as tf
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization

# Use the default TextVectorization parameters
text_vectorizer = TextVectorization(max_tokens=10000, # how many words in the vocabulary (default is None)
                                    standardize="lower_and_strip_punctuation",
                                    split="whitespace",
                                    ngrams=None, # create groups of n-words?
                                    output_mode="int", # how to map tokens to numbers
                                    output_sequence_length=None, # how long do you want sequences to be?
                                    pad_to_max_tokens=True)
```

```python
train_sentences = [sentence for sentence in train_sentences if isinstance(sentence, str)]
# Remove corresponding label from removed sentence
train_labels = [label for label, sentence in zip(train_labels, train_sentences) if isinstance(sentence, str)]

val_sentences = [sentence for sentence in val_sentences if isinstance(sentence, str)]
val_labels = [label for label, sentence in zip(val_labels, val_sentences) if isinstance(sentence, str)]

print(len(train_sentences), len(val_sentences))
print(len(train_labels), len(val_labels))
```

```python
# Find the average number of tokens (words) in the training tweets
round(sum([len(i.split()) for i in train_sentences])/len(train_sentences))
```

```python
# Setup text vectorization variables
max_vocab_length = 10000 # max number of words to have in our vocabulary
max_length = 15 # max length our sequences will be (e.g. how many wordss from a Tweet does a model see?)

text_vectorizer = TextVectorization(max_tokens=max_vocab_length,
                                    output_mode="int",
                                    output_sequence_length=max_length)
```

```python
# Fit the text vectorizer to the training text
text_vectorizer.adapt(train_sentences)
```
